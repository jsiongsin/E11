```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("Upper_Hearst_Garage.csv")
data2 = pd.read_csv("Lower_Sproul.csv")
data3 = pd.read_csv("Glade.csv")
data4 = pd.read_csv("Berkeley_Way_West.csv")
data5 = pd.read_csv("Bancroft_Library.csv")
meta_data = [data1, data2, data3, data4, data5]
```

```python
for data in meta_data:
    plt.hist(data["counts"], bins=15, alpha=0.7)
    plt.title("Radiation")
    plt.ylabel("Counts")
    plt.legend(["Upper Hearst Garage", "Lower Sproul", "Glade", "Berkeley Way West", "Bancroft_Library"], loc="upper right")
    
print("Mean =", np.mean(data["counts"]), "Standard Deviation =", np.std(data["counts"]))
```

```python
for data in meta_data:
    plt.hist(data["Temperature"], bins=11, alpha=0.7)
    plt.title("Temperature")
    plt.ylabel("Temperature")
    plt.legend(["Upper Hearst Garage", "Lower Sproul", "Glade", "Berkeley Way West", "Bancroft_Library"], loc="upper right")
    
print("Mean =", np.mean(data["Temperature"]), "Standard Deviation =", np.std(data["Temperature"]))
```

```python
for data in meta_data:
    plt.hist(data["PM25"], bins=25, alpha=0.7)
    plt.title("Air Quality: 2.5 PM Concentration")
    plt.ylabel("Parts per Million")
    plt.legend(["Upper Hearst Garage", "Lower Sproul", "Glade", "Berkeley Way West", "Bancroft_Library"], loc="upper right")
    
print("Mean =", np.mean(data["PM25"]), "Standard Deviation =", np.std(data["PM25"]))
```

```python
for data in meta_data:
    plt.hist(data["Altitude"], bins=15, alpha=0.7)
    plt.title("Altitude")
    plt.ylabel("Meters")
    plt.legend(["Upper Hearst Garage", "Lower Sproul", "Glade", "Berkeley Way West", "Bancroft_Library"], loc="upper right")
    
print("Mean =", np.mean(data["Altitude"]), "Standard Deviation =", np.std(data["Altitude"]))
```

```python
total_counts = []
altitude = []

for data in meta_data: 
    total_counts.append(np.sum(data["counts"]))

for data in meta_data: 
    altitude.append(np.mean(data["Altitude"]))
```

```python
plt.scatter(altitude, total_counts)
plt.title("Altitude and Radiation Correlation")
plt.ylabel("Counts")
```
